```python
#TASK 1

# Create the years and durations lists
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

# Create a dictionary with the two keys
movie_dict = {"years" : years, "durations" : durations}

# Print the dictionary
movie_dict
```

```python
#TASK 2

# Import pandas using its usual alias
import pandas as pd

# Create a DataFrame from the dictionary
durations_df = pd.DataFrame(movie_dict)

# Print the entire DataFrame
print(durations_df.head)
```

```python
#TASK 3

# Import matplotlib.pyplot under the alias plt
import matplotlib.pyplot as plt
fig = plt.figure()

# Create a line plot of Duration vs Release Years
plt.plot(durations_df["years"], durations_df ["durations"])
plt.xlabel("Release Years")
plt.ylabel("Durations")

# Add a title
plt.title("Netflix Movie Durations 2011-2020")

# Show the plot
plt.show()
```

```python
#TASK 4

# Create DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Print the first five rows of the DataFrame
print(netflix_df.head(5))
```

```python
#TASK 5

# Subset the DataFrame above for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df["type"] == "Movie"]

# Subset only the column of interest 
netflix_movies_col_subset = netflix_df_movies_only[["title", "country", "genre", "release_year", "duration"]]

# Print the first five rows of the new DataFrame
print(netflix_movies_col_subset.head(5))
```

```python
#TASK 6

# Create a scatter plot
fig = plt.figure()

# Create a scatter plot of Duration vs Release Year
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"])
plt.xlabel("Release Year")
plt.ylabel("Duration")

# Add a title
plt.title("Movies Duration by Year of Release")

# Show the plot
plt.show()
```

```python
#TASK 7

# Subset to create only movies fewer than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"] < 60]

# Print the first 20 rows
print(short_movies.head(20))
```

```python
#TASK 8

# Initialize an empty list
colors = []

# Iterate 
for lab, row in netflix_movies_col_subset.iterrows() : 
    if row["genre"] == "Children" : 
        colors.append("red")
    elif row["genre"] == "Documentaries" : 
        colors.append("blue")
    elif row["genre"] == "Stand-Up" :
        colors.append("green")
    else:
        colors.append("black")

# Print the first 10 rows 
print(colors[:10])
```

```python
TASK 9

# Create a scatter plot
plt.style.use("fivethirtyeight")
fig = plt.figure()

# Create a scatter plot of Duration vs Release Year
plt.scatter(netflix_movies_col_subset["release_year"],netflix_movies_col_subset["duration"],c=colors)

# Add a title and axis label
plt.xlabel("Release Year")
plt.ylabel("Duration(min)")
plt.title("Movie Duration by Year of Release")

# Show the plot
plt.show()
```
